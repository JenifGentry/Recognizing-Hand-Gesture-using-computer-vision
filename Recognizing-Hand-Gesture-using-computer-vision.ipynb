#**Image Augumentation:**

The Image Augumentation is done in order to add complex images of different type so that the model can be trained on more complex images. This process will read the image and adjust contrast of the image (Budka, M., 2023), flip the image and darker the image randomly.

```python
#Adjust the contrast of the image
def contrast(im, min_alpha=0.6, max_alpha=1.4):
    alpha = np.random.uniform(min_alpha, max_alpha)
    beta = 100
    im_tfm = np.clip(alpha * im + beta, 0, 255).astype('uint8')
    return im_tfm
#Flip the image
def fliplr(im):
  return im[:, ::-1]
#Darker the image
def darker(im):
  return (0.5*im).astype(im.dtype)
#Randomly choose image
def rand_tfm(im, tfm, p=0.5):
    if np.random.random() < p:
        return tfm(im)
    return im
```

#**Training Model:**
In the fit function the model is trained using the training and testing datasets and control various training parameters such as learning rate, weight decay, and optimizer choice. The training and testing accuracy is displayed.

```python
def fit(model, train_data, test_data, criterion, epochs, lr, weight_decay,
        optimizer=torch.optim.Adam,
         device=torch.device("cuda:0" if torch.cuda.is_available() else "cpu"),
        plot=True):

    # Move the model to the specified device
    model.to(device)

    # Define the optimizer
    optimizer = optimizer(model.parameters(), lr=lr, weight_decay=weight_decay)

    # Initialize the lists to store losses and accuracies
    train_loss_hist, train_acc_hist = [], []
    test_loss_hist, test_acc_hist = [], []

    for epoch in range(epochs):
        # Training phase
        model.train()
        train_loss, train_acc = 0.0, 0.0
        for i, (inputs, labels) in enumerate(train_data):
            inputs = inputs.to(device)
            labels = labels.to(device)

            optimizer.zero_grad()

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            train_acc += torch.sum(preds == labels.data)

        epoch_train_loss = train_loss / len(train_data.dataset)
        epoch_train_acc = train_acc.double() / len(train_data.dataset)
        train_loss_hist.append(epoch_train_loss)
        train_acc_hist.append(epoch_train_acc)

        # Validation phase
        model.eval()
        test_loss, test_acc = 0.0, 0.0
        with torch.no_grad():
            for i, (inputs, labels) in enumerate(test_data):
                inputs = inputs.to(device)
                labels = labels.to(device)

                outputs = model(inputs)
                loss = criterion(outputs, labels)

                test_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                test_acc += torch.sum(preds == labels.data)

            epoch_test_loss = test_loss / len(test_data.dataset)
            epoch_test_acc = test_acc.double() / len(test_data.dataset)
            test_loss_hist.append(epoch_test_loss)
            test_acc_hist.append(epoch_test_acc)

        # Print progress
        print(f"Epoch {epoch+1}/{epochs}, Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.4f}, Test Loss: {epoch_test_loss:.4f}, Test Acc: {epoch_test_acc:.4f}")

    # Plot losses
    if plot:
        _, ax = plt.subplots(1, 1, figsize=(10, 6))
        ax.plot(range(1, epochs+1), train_loss_hist, label='train')
        ax.plot(range(1, epochs+1), test_loss_hist, label='test')
        ax.set_xlabel('Epochs')
        ax.set_ylabel('Loss')
        ax.legend()
        plt.show()

    #return train_loss_hist, train_acc_hist, test_loss_hist, test_acc_hist
```

#**Building ResNet model:**
The ResNet-50 model is initialised with pretrained weights in this step, and then the fit function is called to train the model using the provided training and testing datasets.

```python
resnet50_model = torchvision.models.resnet50(pretrained=True)
num_features = resnet50_model.fc.in_features
resnet50_model.fc = nn.Linear(num_features, 8)
fit(resnet50_model, train_data, test_data, criterion=torch.nn.CrossEntropyLoss(), epochs=5, lr=0.0001,weight_decay=1, plot=True)
```

#**Building AlexNet model**
Pretrained weights are used to initialise the AlexNet model in this stage, after which the fit function is called to train the model using the provided training and testing datasets. Dropout layer is used in place of Alexnet's last layer in this instance.

```python
# Load the pre-trained AlexNet model
alexnet_model= torchvision.models.alexnet(pretrained=True)

# Freeze the convolutional layers
for param in alexnet_model.features.parameters():
    param.requires_grad = False

# Replace the last fully connected layer with a new one
num_ftrs = alexnet_model.classifier[6].in_features
alexnet_model.classifier[6] = nn.Sequential(
    nn.Dropout(p=0.8),
    nn.Linear(num_ftrs, 8)
)
fit(alexnet_model, train_data, test_data, criterion=nn.CrossEntropyLoss(), epochs=5, lr=0.0001, weight_decay=0.001)
```

#**Building CNN model:**
In this step, CNN model is created and then calls the fit function to train the model using the specified training and testing datasets.

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dropout = nn.Dropout(p=0.8)
        self.fc1 = nn.Linear(64 * 56 * 56, 128)
        self.fc2 = nn.Linear(128, 8)

    def forward(self, x):
        x = self.pool(nn.functional.relu(self.conv1(x)))
        x = self.pool(nn.functional.relu(self.conv2(x)))
        x = x.view(-1, 64 * 56 * 56)
        x = self.dropout(nn.functional.relu(self.fc1(x)))
        x = self.fc2(x)
        return x
cnn_model = CNN()
fit(cnn_model, train_data, test_data, criterion=nn.CrossEntropyLoss(), epochs=5, lr=0.001, weight_decay=0.01)
```

In order to improve the performance of the ResNet50 model, the last layer of resnet50 is replaced with batch normalization.

```python
resnet50_model_BatchNorm = torchvision.models.resnet50(pretrained=True)
num_ftrs = resnet50_model_BatchNorm.fc.in_features
resnet50_model_BatchNorm.fc = nn.Sequential(nn.BatchNorm1d(num_ftrs),
    nn.Linear(num_ftrs, 8)
)
fit(resnet50_model_BatchNorm, train_data, test_data, criterion=torch.nn.CrossEntropyLoss(), epochs=5, weight_decay=1, lr=0.0001, plot=True)
```

In order to capture intermediate outputs during a forward pass or gradients during a backward pass on a PyTorch module, this class offers a straightforward mechanism to register hooks on a PyTorch module. The out attribute of the RegisterHook instance provides access to the outputs or gradients that were captured (Budka, M., 2023).

```python
class RegisterHook(object):

    out = None

    def get_hook_fn(self):
        def fn(m, inp, out): self.out = out
        return fn

    def __init__(self, m, fwd=True):
        self.m, self.fwd = m, fwd

    def __enter__(self):
        hook_fn = self.get_hook_fn()
        if self.fwd: self.h = self.m.register_forward_hook(hook_fn)
        else:        self.h = self.m.register_backward_hook(hook_fn)
        return self

    def __exit__(self, type, value, traceback):
        self.h.remove()
```

The denorm function takes an input tensor and denormalizes it using either the specified statistics or, in the absence of any statistics, the default ImageNet statistics. This is frequently used to transform normalised image data in computer vision tasks.

```python
imagenet_stats = ([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
def denorm(x, stats=imagenet_stats):
    return x * torch.Tensor(stats[1])[:,None,None] + torch.Tensor(stats[0])[:,None,None]
```

The original input image x, the target class y, and the heatmap produced by the occlude function are all inputs to the show_heatmap function. The heatmap is displayed above the input image, with a colormap used to show the values in the heatmap. The areas of the input image that are most crucial for class prediction are visible in the final image.

```python
def occlude(x, y, model,occ_sz=32, occ_stride=16):
    h,w = x.shape[1:]
    hm = []
    for r in trange(0, w, occ_stride-1):
        hm_r = []
        for c in range(0, h, occ_stride-1):
            x_cpy = x.clone()
            x_cpy[:, r:r+occ_sz, c:c+occ_sz] = 0.5
            out = model(x_cpy[None,...].cuda())
            hm_r.append(out[0,y].item())
        hm.append(hm_r)
    return np.array(hm)

def show_heatmap(x, y, hm):
    ax = show_img(denorm(x).numpy().transpose(1,2,0), title=im_labels[y])
    ax.imshow(hm, alpha=0.5, extent=(0,x.shape[2],x.shape[1],0), interpolation='bilinear', cmap='hot')
```

A certain number of accurately predicted images from a given model are shown using the show_correct_predictions function.

```python
def show_correct_predictions(num_images, model, figsize=(8, 8)):
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    model.eval() # set model to evaluation mode
    with torch.no_grad(): # turn off gradient calculation to save memory
        correct = 0
        for i, (images, labels) in enumerate(test_data): # loop through test data loader
            images, labels = images.to(device), labels.to(device) # move tensors to the device
            outputs = model(images) # forward pass
            _, predicted = torch.max(outputs, 1) # get predicted labels
            for j in range(images.shape[0]): # loop through images in batch
                if np.array_equal(predicted[j].cpu().numpy(), labels[j].cpu().numpy()): # if prediction is correct
                    img = images[j].cpu().clone()
                    true_label = im_labels[labels[j].cpu().numpy()] # get true label
                    predicted_label = im_labels[predicted[j].cpu().numpy()] # get predicted label
                    print(f"True label: {true_label}, Predicted label: {predicted_label}")
                    hm = occlude(img, labels[j], occ_sz=16, occ_stride=8, model=model)
                    with RegisterHook(model.layer4, fwd=True) as fhk:
                      model(img[None,...].cuda())
                      hm = fhk.out.mean(dim=1).detach().squeeze().cpu().numpy()
                      # show_heatmap(x2,y2,hm)
                      show_heatmap(img, labels[j], hm)
                    correct += 1
                if correct == num_images:
                    break
            if correct == num_images:
                break
```

The incorrect_images function is used to find and show images from a given model that were incorrectly predicted (Budka, M., 2023).

```python
def incorrect_images(model):
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    model.eval() # set model to evaluation mode
    faulty_images = []
    with torch.no_grad(): # turn off gradient calculation to save memory
        for i, (images, labels) in enumerate(test_data): # loop through test data loader
            images, labels = images.to(device), labels.to(device) # move tensors to the device
            outputs = model(images) # forward pass
            _, predicted = torch.max(outputs, 1) # get predicted labels
            for j in range(images.shape[0]): # loop through images in batch
                if not np.array_equal(predicted[j].cpu().numpy(), labels[j].cpu().numpy()): # if prediction is correct


                    img = images[j].cpu().numpy().transpose((1, 2, 0)) # get image
                    true_label = im_labels[labels[j].cpu().numpy()] # get true label
                    predicted_label = im_labels[predicted[j].cpu().numpy()] # get predicted label
                    print(f"True label: {true_label}, Predicted label: {predicted_label}")
                    plt.imshow(img) # display image
                    plt.show() # show image
```

```python
def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
```
